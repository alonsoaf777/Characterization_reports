```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
peaks_predict = pd.read_csv('Predict_unknown3_processed.xy', delim_whitespace=True)
peaks_predict.columns = ['2Theta', 'I']
df_lattices = pd.read_csv('real_angles.txt', delim_whitespace=True)
peaks_predict2 = pd.read_csv('Predict_unknown4_processed.xy', delim_whitespace=True)
peaks_predict2.columns = ['2Theta', 'I']
df_lattices2 = pd.read_csv('1100138.txt', delim_whitespace=True)
#Take the important angles
peaks = [0,1,2,3,4,6,7,8,9,10,11,12,13,14,18,19]
important_lattices = df_lattices.loc[peaks]
important_lattices = important_lattices.reset_index(drop=True)
```

```python
lattices = [(1,0,1), (1,0,3), (0,0,4), (1,1,2), (2,0,0), (1,0,5), (2,1,1), (2,1,3), (2,0,4), (1,1,6), (2,2,0), (1,0,7), (2,1,5), (3,0,1), (2,2,4), (3,1,2)]
```

```python
# Unir las listas en una lista de tuplas
x_valuesr = important_lattices['2Theta'].tolist()
y_valuesr = important_lattices['I'].tolist()
intensidades = [105, 30, 40, 30, 60, 30, 40, 15, 30, 10, 10, 50, 50, 50, 10, 5]
angules_ =     [0, 45, -45, -45, 45, 45, -45, 45, -45, 45, 45,45,45,45,45,-45]
data_points = list(zip(x_valuesr, y_valuesr))
fig = plt.Figure()
aux = 0

for angulo in important_lattices['2Theta']:
    if aux == 0:
        plt.axvline(x=angulo, color='k', linestyle='--', ymax=10, label='Ángulo teórico', linewidth = 0.3)
        
    else:
        plt.axvline(x=angulo, color='k', linestyle='--', ymax=110, linewidth = 0.3)
        
    aux += 1

aux = 0
for x, y in data_points:
        if aux == 4:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 2, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 1:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 2, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 5:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 2, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 7:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 2, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 9:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 2, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 10:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 1, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 11:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 1, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 13:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 1, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        elif aux == 14:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x - 1, intensidades[aux]-3),
                    horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        else:
            plt.annotate(f'{lattices[aux]}', xy=(x, y), xytext=(x + 2, intensidades[aux]-3),
            horizontalalignment='center', rotation=angules_[aux], fontsize = 6)
        aux += 1

plt.plot(peaks_predict['2Theta'], peaks_predict['I'], color = 'blue', linewidth = .4, label = 'Experimental')
plt.xlabel('Bragg 2θ',fontsize=12)
plt.ylabel('Intensidad',fontsize=12)
plt.ylim(0, 110)
plt.title('Picos de Difracción TiO2', fontsize=16)
plt.legend(fontsize = 8)
plt.show()
```
